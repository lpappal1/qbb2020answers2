```python
import numpy as np
import matplotlib as plt

ER4 = open('/Users/cmdb/qbb2020-answers/week5/g1e/ER4_overlaps.bed', 'r')
g1e = open('/Users/cmdb/qbb2020-answers/week5/g1e/g1e_overlaps.bed', 'r')

g1e_introns = 0
g1e_exons = 0
g1e_promoters = 0
ER4_introns = 0
ER4_exons = 0
ER4_promoters = 0

for i in ER4:
    line = i.rstrip('\n\r').split('\t')
    if "promoter" in line:
        ER4_promoters += 1
    if "exon" in line:
        ER4_exons += 1
    if "intron" in line:
        ER4_introns += 1
        
for i in g1e:
    line = i.rstrip('\n\r').split('\t')
    if "promoter" in line:
        g1e_promoters += 1
    if "exon" in line:
        g1e_exons += 1
    if "intron" in line:
        g1e_introns += 1
    
print(ER4_promoters)
print(ER4_exons)

#f.close()

```

```python
#Using the + and - of the mouse vs. ER4 to find losses/gains
ER4 = open('/Users/cmdb/qbb2020-answers/week5/g1e/ER4_overlaps.bed', 'r')
g1e = open('/Users/cmdb/qbb2020-answers/week5/g1e/g1e_overlaps.bed', 'r')

maybegains = 0
maybelosses = 0
g1elosses = 0
g1egains = 0

for i in ER4:
    line = i.rstrip('\n\r').split('\t')
    if "+" in line:
        maybegains += 1
    if "-" in line:
        maybelosses += 1
        
for i in g1e:
    line = i.rstrip('\n\r').split('\t')
    if "+" in line:
        g1egains += 1
    if "-" in line:
        g1elosses += 1
```

```python
exons = [g1e_exons, ER4_exons]
introns = [g1e_introns, ER4_introns]
promoters = [g1e_promoters, ER4_promoters]
print(exons)
print(introns)
print(promoters)
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1,2)

objects = ('ER4_prom', 'ER4_exon', "ER4_intron", "g1e_prom", "g1e_exon", "g1e_intron")
y_pos = np.arange(len(objects))
value = [ER4_promoters, ER4_exons, ER4_introns, g1e_promoters, g1e_exons, g1e_introns]
ax[0].bar(y_pos, value)
ax[0].set_ylabel('Number of sites')
ax[0].set_xlabel('DNA Type')
ax[0].set_title('CTCF Sites by DNA Type')
ax[0].set_xticks(y_pos)
ax[0].set_xticklabels(objects, rotation='vertical')

objects1 = ('ER4_loss', 'ER4_gain', "g1e_loss", "g1e_gain")
y_pos1 = np.arange(len(objects1))
value1 = [maybelosses, maybegains, g1elosses, g1egains]

ax[1].bar(y_pos1, value1, align='center', alpha=0.5)
ax[1].set_xticks(y_pos1)
ax[1].set_xticklabels(objects1, rotation='vertical')

ax[1].set_ylabel('Number of Sites')
ax[1].set_xlabel('Cell Type')
ax[1].set_title('CTCF Site Changes')

#ax[1].ylabel('Counts')
ax[1].set_title('Gains and Losses')
```
